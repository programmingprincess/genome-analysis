```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
samples=['043F13','043F14','043F15','045F14','045F15','045F16']
```

```python
# left is coverage 
# right is high quality coverage count 
x = np.loadtxt('043F13-recal.txt', skiprows=11)
```

```python
for sample in samples:
    x = np.loadtxt(sample+'-recal.txt', skiprows=11)
    bins = x[:,0]
    counts = x[:,1]
    plt.bar(bins, counts)
    plt.title(sample+".bam")
    plt.xlabel("Coverage")
    plt.ylabel("HQ Coverage Count")
    plt.show()
```
